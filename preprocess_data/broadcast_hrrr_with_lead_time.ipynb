```python
import xarray as xr
import os
import numpy as np
import pandas as pd

from functools import partial

HRRR_LT_DATA_PATH = "/g/data/wa46/user/nl5316/tw_spatial/hrrr_with_lead_time"
```

```python
file_list = sorted([f for f in os.listdir(HRRR_LT_DATA_PATH) if f.endswith(".nc")])
ncs = file_list
ncs = [HRRR_LT_DATA_PATH + "/" + nc for nc in ncs]
```

```python
ncs[0]
```

```python
ds = xr.open_dataset(ncs[0])
```

```python
for nc, file_name in zip(ncs, file_list):
    ds = xr.open_dataset(nc)
    broadcast_array = xr.DataArray(
        [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan],
        dims=["lead_time"],
        coords={"lead_time": pd.timedelta_range(start="6h", end="2d", freq="6h").values},
    )
    ds = ds.broadcast_like(broadcast_array)
    ds.to_netcdf(f"/g/data/wa46/user/nl5316/tw_spatial/hrrr_combined/{file_name}")
    print(f"broadcast data for {file_name}")
```
